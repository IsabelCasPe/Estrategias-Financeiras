```python
import random
import matplotlib.pyplot as plt
```

```python
# Lista de 5 países
paises = ["Brasil", "Estados Unidos", "Japão", "Alemanha", "Índia"]

# Gerando PIP fictício para cada país (em bilhões de dólares)
pip_ficticio = {pais: random.randint(500, 5000) for pais in paises}

# Exibindo os resultados
print("PIP Fictício dos Países (em bilhões de dólares):")
for pais, pip in pip_ficticio.items():
    print(f"{pais}: ${pip} bilhões")
```

```python
# Lista de 5 países
paises = ["Brasil", "Estados Unidos", "Japão", "Alemanha", "Índia"]

# Gerando PIP fictício para cada país (em bilhões de dólares)
pip_ficticio = {pais: random.randint(500, 5000) for pais in paises}

# Exibindo os resultados
print("PIP Fictício dos Países (em bilhões de dólares):")
for pais, pip in pip_ficticio.items():
    print(f"{pais}: ${pip} bilhões")

# Criando o gráfico de barras
plt.bar(pip_ficticio.keys(), pip_ficticio.values(), color='skyblue')

# Adicionando detalhes ao gráfico
plt.title("PIP Fictício dos Países (em bilhões de dólares)")
plt.xlabel("Países")
plt.ylabel("PIP (em bilhões de dólares)")
plt.xticks(rotation=45)  # Rotaciona os nomes dos países para melhor visualização
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adiciona uma grade para facilitar a leitura

# Exibindo o gráfico
plt.tight_layout()  # Ajusta o layout para evitar cortes
plt.show()
```
