```python
# Reimportar bibliotecas ap√≥s o reset
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir os par√¢metros da calculadora
def calcular_juros_compostos(principal, taxa_juros, periodos):
    """
    Calcula o montante final com juros compostos.
    """
    montante = principal * (1 + taxa_juros) ** periodos
    return montante

# Configura√ß√µes do exemplo
principal = 1000  # Investimento inicial (em unidades monet√°rias)
taxa_juros = 0.05  # Taxa de juros (5% ao per√≠odo)
periodos = 10  # N√∫mero de per√≠odos

# Calcular o montante final
montante_final = calcular_juros_compostos(principal, taxa_juros, periodos)

# Gerar dados para o gr√°fico
periodos_array = np.arange(1, periodos + 1)
montantes = [calcular_juros_compostos(principal, taxa_juros, t) for t in periodos_array]

# Criar o gr√°fico
plt.figure(figsize=(8, 5))
plt.plot(periodos_array, montantes, marker='o', label="Montante")
plt.title("Crescimento com Juros Compostos")
plt.xlabel("Per√≠odos")
plt.ylabel("Montante (Unidades Monet√°rias)")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

montante_final
```
