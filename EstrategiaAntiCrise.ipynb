```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import mplcyberpunk
plt.style.use("cyberpunk")
```

Passo 2 - Pagar os dados do yahoo

```python
ticker = "ALPA4.SA"
```

```python
dados = yf.download(ticker, start="2021-05-27", end = "2024-09-10")
dados
```

```python
dados = dados.droplevel(1, axis=1)
dados
```

Passo 3 - Definir as janelas das médias móveis

```python
janela_rapida = 7
janela_lenta = 40
```

Passo 4 - Calcular as médias móveis

```python
dados["media_rapida"] =  dados["Adj Close"].rolling(janela_rapida).mean()
dados["media_lenta"] =  dados["Adj Close"].rolling(janela_lenta).mean()
dados
```

```python
dados.head(50)
```

Passo 5 - Calcular o retorno diário da ação

```python
dados["retorno_diario"] = dados["Adj Close"].pct_change()
dados = dados.dropna()
dados
```

Passo 6 - Criar os sinais de compra ou venda

```python
dados["camprado_ou_vendido"] = 0
```

```python
dados["camprado_ou_vendido"] = np.where(dados["media_rapida"] > dados["media_lenta"], 1, -1)
dados
```

```python
dados["camprado_ou_vendido"] = dados["camprado_ou_vendido"].shift(1)
dados
```

Passo 7 - Calcular o retorno do modelo

```python
dados["retorno_modelo"] = dados["camprado_ou_vendido"] * dados["retorno_diario"]
dados
```

```python
dados["retorno_acum_modelo"] = (1 + dados["retorno_modelo"]).cumprod()-1
dados
```

Passo 8 - Calcular o retorno da ação e do Ibovespa np mesmo período

```python
dados["retorno_acum_acao"] = (1 + dados["retorno_diario"]).cumprod()-1
dados
```

```python
ibov=yf.download("^BVSP", start = dados.index[0], end = "2024-09-10")
ibov = ibov.droplevel(1, axis=1)
ibov = ibov["Close"]
ibov
```

```python
dados["retorno_ibov"] = ibov.pct_change().values
dados["retorno_acum_ibov"] = (1 + dados["retorno_ibov"]).cumprod()-1
dados = dados.dropna()
dados
```

Passo 9 - Criar o gráfico de retornos para observar

```python
plt.figure(figsize=(14,7))
plt.plot(dados.index, dados["retorno_acum_modelo"], label = "Estrategia")
plt.plot(dados.index, dados["retorno_acum_acao"], label = "Acao")
plt.plot(dados.index, dados["retorno_ibov"], label = "IBOV")
plt.legend()
plt.show()
```

Passo 10 - Otimizar os resultados

```python
intervalo_rapido = range(5, 30, 1)  # Janela rápida: 5 a 30 dias
intervalo_lento = range(30, 100, 5)  # Janela lenta: 30 a 100 dias
```

```python
melhor_retorno = -np.inf  # Inicializa o melhor retorno
melhores_parametros = None

# Itera pelos intervalos de médias móveis rápidas e lentas
for janela_rapida in intervalo_rapido:
    for janela_lenta in intervalo_lento:
        if janela_rapida >= janela_lenta:
            continue  # Evita combinações inválidas

        # Recalcula as médias móveis
        dados["media_rapida"] = dados["Adj Close"].rolling(janela_rapida).mean()
        dados["media_lenta"] = dados["Adj Close"].rolling(janela_lenta).mean()

        # Gera os sinais de compra/venda
        dados["camprado_ou_vendido"] = np.where(dados["media_rapida"] > dados["media_lenta"], 1, -1)
        dados["camprado_ou_vendido"] = dados["camprado_ou_vendido"].shift(1)  # Ajusta para não usar "dados futuros"

        # Calcula o retorno do modelo
        dados["retorno_modelo"] = dados["camprado_ou_vendido"] * dados["retorno_diario"]
        dados["retorno_acum_modelo"] = (1 + dados["retorno_modelo"]).cumprod() - 1

        # Calcula o retorno final
        retorno_final = dados["retorno_acum_modelo"].iloc[-1]  # Último valor do retorno acumulado

        # Atualiza o melhor retorno e os parâmetros
        if retorno_final > melhor_retorno:
            melhor_retorno = retorno_final
            melhores_parametros = (janela_rapida, janela_lenta)

print(f"Melhores parâmetros: Rápida = {melhores_parametros[0]}, Lenta = {melhores_parametros[1]}")
print(f"Melhor retorno acumulado: {melhor_retorno:.2%}")
```

```python
# Recalcula com os melhores parâmetros
dados["media_rapida"] = dados["Adj Close"].rolling(melhores_parametros[0]).mean()
dados["media_lenta"] = dados["Adj Close"].rolling(melhores_parametros[1]).mean()
dados["camprado_ou_vendido"] = np.where(dados["media_rapida"] > dados["media_lenta"], 1, -1)
dados["camprado_ou_vendido"] = dados["camprado_ou_vendido"].shift(1)
dados["retorno_modelo"] = dados["camprado_ou_vendido"] * dados["retorno_diario"]
dados["retorno_acum_modelo"] = (1 + dados["retorno_modelo"]).cumprod() - 1

# Plota o gráfico otimizado
plt.figure(figsize=(14,7))
plt.plot(dados.index, dados["retorno_acum_modelo"], label = "Estratégia Otimizada")
plt.plot(dados.index, dados["retorno_acum_acao"], label = "Ação")
plt.plot(dados.index, dados["retorno_acum_ibov"], label = "IBOV")
plt.legend()
plt.show()
```

```python
resultados = []  # Lista para armazenar os resultados

for janela_rapida in intervalo_rapido:
    for janela_lenta in intervalo_lento:
        if janela_rapida >= janela_lenta:
            continue

        # Recalcula as médias móveis
        dados["media_rapida"] = dados["Adj Close"].rolling(janela_rapida).mean()
        dados["media_lenta"] = dados["Adj Close"].rolling(janela_lenta).mean()
        dados["camprado_ou_vendido"] = np.where(dados["media_rapida"] > dados["media_lenta"], 1, -1)
        dados["camprado_ou_vendido"] = dados["camprado_ou_vendido"].shift(1)
        dados["retorno_modelo"] = dados["camprado_ou_vendido"] * dados["retorno_diario"]
        dados["retorno_acum_modelo"] = (1 + dados["retorno_modelo"]).cumprod() - 1

        retorno_final = dados["retorno_acum_modelo"].iloc[-1]

        # Adiciona o resultado na lista
        resultados.append({
            "Janela Rápida": janela_rapida,
            "Janela Lenta": janela_lenta,
            "Retorno Acumulado (%)": retorno_final * 100
        })

# Converte a lista em DataFrame
df_resultados = pd.DataFrame(resultados)
```
