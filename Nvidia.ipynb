```python
import yfinance as yf

# Puxando dados da NVIDIA
nvidia = yf.Ticker("NVDA")

# Informações básicas
print("Preço atual:", nvidia.history(period="1d")['Close'].iloc[-1])
print("Variação do dia:", nvidia.history(period="1d")['Close'].pct_change().iloc[-1] * 100, "%")

# Mais detalhes
print("\nInformações da empresa:")
print(nvidia.info)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Baixando dados da NVIDIA (NVDA)
nvidia = yf.Ticker("NVDA")

# Puxando dados históricos dos últimos 6 meses
dados = nvidia.history(period="6mo")

# Criando gráficos
plt.figure(figsize=(14, 8))

# Gráfico 1: Preço de Fechamento
plt.subplot(2, 2, 1)  # 2 linhas, 2 colunas, posição 1
plt.plot(dados['Close'], label='Preço de Fechamento', color='blue')
plt.title('Preço de Fechamento da NVIDIA (6 meses)')
plt.xlabel('Data')
plt.ylabel('Preço (USD)')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)

# Gráfico 2: Volume Negociado
plt.subplot(2, 2, 2)  # 2 linhas, 2 colunas, posição 2
plt.bar(dados.index, dados['Volume'], color='green', alpha=0.7)
plt.title('Volume Negociado da NVIDIA (6 meses)')
plt.xlabel('Data')
plt.ylabel('Volume')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 3: Variação Diária
plt.subplot(2, 2, 3)  # 2 linhas, 2 colunas, posição 3
variacao_diaria = dados['Close'].pct_change() * 100
plt.plot(variacao_diaria, label='Variação Diária (%)', color='red')
plt.title('Variação Diária da NVIDIA (6 meses)')
plt.xlabel('Data')
plt.ylabel('Variação (%)')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)

# Gráfico 4: Média Móvel de 30 dias
plt.subplot(2, 2, 4)  # 2 linhas, 2 colunas, posição 4
dados['MA_30'] = dados['Close'].rolling(window=30).mean()
plt.plot(dados['Close'], label='Preço de Fechamento', color='blue', alpha=0.5)
plt.plot(dados['MA_30'], label='Média Móvel (30 dias)', color='orange')
plt.title('Média Móvel de 30 dias da NVIDIA')
plt.xlabel('Data')
plt.ylabel('Preço (USD)')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)

# Ajustando o layout
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```
