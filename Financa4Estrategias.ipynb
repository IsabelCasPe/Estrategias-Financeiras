


1. Internacionalizar uma Companhia

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simular receitas de diferentes regies
np.random.seed(42)
dias = 365  # Perodo de 1 ano
mercados = ['Amrica do Norte', 'Europa', 'sia', 'Amrica Latina']

# Gerar receitas simuladas (em milhes) para cada mercado
receitas = {
    mercado: np.cumsum(np.random.normal(5, 2, dias)) for mercado in mercados  # Crescimento dirio
}

# Criar DataFrame com receitas acumuladas
df_receitas = pd.DataFrame(receitas, index=pd.date_range(start="2024-01-01", periods=dias))

# Calcular receitas totais e diversificao
df_receitas['Total'] = df_receitas.sum(axis=1)
df_receitas_percentual = df_receitas.div(df_receitas['Total'], axis=0) * 100

# Plotar as receitas acumuladas por mercado
plt.figure(figsize=(12, 6))
for mercado in mercados:
    plt.plot(df_receitas.index, df_receitas[mercado], label=mercado)
plt.plot(df_receitas.index, df_receitas['Total'], label='Total', color='black', linewidth=2)
plt.title("Receitas Acumuladas por Regio")
plt.xlabel("Data")
plt.ylabel("Receitas (milhes)")
plt.legend()
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# Mostrar porcentagens de participao final de cada mercado
participacao_final = df_receitas_percentual.iloc[-1, :-1]
print("Participao Final de Receitas por Mercado:")
print(participacao_final)
```

```python
# Simular dados financeiros da empresa
capital_proprio = 500000  # Capital prprio (em milhares)
lucro_liquido = 80000  # Lucro lquido (em milhares)
divida_inicial = 200000  # Dvida inicial (em milhares)

# Clculo da Razo Dvida/Capital Prprio (D/E Ratio)
de_ratio = divida_inicial / capital_proprio

# Clculo do ROE (antes e depois de aumentar a dvida)
roe_inicial = lucro_liquido / capital_proprio

# Simular diferentes nveis de dvida e recalcular o ROE
divida_simulada = np.arange(200000, 600000, 50000)  # Dvida variando de 200k a 600k
roe_simulado = [lucro_liquido / (capital_proprio + divida) for divida in divida_simulada]

# Criar DataFrame com os resultados
df_divida_roe = pd.DataFrame({
    "Dvida (milhares)": divida_simulada,
    "Razo D/E": divida_simulada / capital_proprio,
    "ROE (%)": np.array(roe_simulado) * 100
})

# Plotar o impacto da dvida no ROE
plt.figure(figsize=(10, 6))
plt.plot(df_divida_roe["Dvida (milhares)"], df_divida_roe["ROE (%)"], marker='o', color='blue', label="ROE (%)")
plt.title("Impacto da Dvida no ROE")
plt.xlabel("Dvida (milhares)")
plt.ylabel("ROE (%)")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
